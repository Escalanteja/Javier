For the missing first_name values we have about 11,259 of those values missing. This will not make much of a difference on our data analysis because we have the user_id column which also helps us identify customers in a numerical format so we can leave the missing value for first_name

```python
# check for mixed data types
for col in customers.columns.tolist():
  weird = (customers[[col]].applymap(type) != customers[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (customers[weird]) > 0:
    print (col)
```

# first_name has mixed data type

```python
# Convert first_name data type to string
customers['first_name'] = customers['first_name'].astype('str')
```

```python
# data type check
customers.dtypes
```

```python
# check again for any other mixed data types
for col in customers.columns.tolist():
  weird = (customers[[col]].applymap(type) != customers[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (customers[weird]) > 0:
    print (col)
```

There are no other mixed data types
